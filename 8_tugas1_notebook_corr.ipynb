Seleksi Fitur

```python
import seaborn as sns
import matplotlib.pyplot as plt

selected_features = [
    "Energy (Per 100 ml)",
    "Fat (Per 100 ml)",
    "Saturated fat (Per 100 ml)",
    "Carbohydrates (Per 100 ml)",
    "Sugars (Per 100 ml)",
    "Fiber (Per 100 ml)",
    "Proteins (Per 100 ml)",
    "Salt (Per 100 ml)"
]

corrmat = df[selected_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corrmat, annot=True, cmap="RdYlGn", fmt=".2f")
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

korelasi tinggi antar fitur
1. fat dan saturated fat = 0.97
2. energy dan fat = 0.76
3. energy dan saturated fat = 0.70

korelasi sedang
1. carbohydrates dan sugars = 0.70
2. energy dan carbohydrates = 0.65
3. carbohydrates dan salt = 0.45
4. energy dan proteins = 0.47
